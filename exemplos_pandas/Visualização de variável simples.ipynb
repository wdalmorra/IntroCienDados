```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('zap.csv',encoding='utf-8')
```

```python
len(df)
```

```python
df['preco_int'] = df.preco.apply(lambda x:int(x.replace(".","")))
```

```python
df.groupby('bairro').count().preco.sort_values(ascending=False).head()
```

```python
df.preco_int.describe()
```

```python
import seaborn as sns

(df.preco_int/1000).plot(kind='hist',bins=200)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(df[(df.bairro=='Centro') & (df.preco_int<=1000000)].preco_int/1000,bw=20)
sns.kdeplot(df[(df.bairro== u'Areal') & (df.preco_int<=1000000)].preco_int/1000,bw=20)
plt.ylim(0,0.004)
plt.grid()
```

```python
plt.boxplot([(df[(df.bairro=='Centro') & (df.preco_int<=1000000)].preco_int/1000),(df[(df.bairro=='Areal') & (df.preco_int<=1000000)].preco_int/1000)])
```
